```python
from datetime import datetime
from time import time
import pandas as pd
import os, re
import numpy as np
```

```python
present = time()
```

```python
fn_format = '%Y-%m-%d %H-%M-%S'
```

```python
pattern = re.compile(r'\d{4}-\d{2}-\d{2} \d{2}-\d{2}-\d{2}')
```

```python
files = os.listdir('./data')
files_str = ' '.join(files)
files_list = re.findall(pattern, files_str)
```

```python
files_list[:10]
```

```python
files_timestamp = []
```

```python
for each in files_list:
    files_timestamp.append(datetime.strptime(each, fn_format).timestamp())
```

```python
files_timestamp[:10]
```

```python
temp = dict(zip(files_timestamp, files_list))
for key, value in temp.items():
    print(key, value)
```

```python
files_timestamp.sort(reverse=True)
```

```python
for i in files_timestamp:
    print(i)
```
